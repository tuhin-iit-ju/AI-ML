C. Even, Odd, Positive and Negative

```python
a = int(input())
b = input().split()
e,o,p,n=0,0,0,0
for i in range(a):
    t= int(b[i])
    if t%2==0 :
        e+=1

    else :
        o+=1
    if t> 0 :
        p+=1
    if t<0 :
        n+=1
print("Even:", e)
print("Odd:", o)
print("Positive:", p)
print("Negative:", n)
 

```
